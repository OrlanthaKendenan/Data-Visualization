# Analysis

## A) Billionaires

#### 1. Determine the distribution by gender and self-made

```python
#fig, (ax1,ax2) = plt.subplots(1,2,figsize=(20,10))

#Pie chart
color = ['darksalmon','purple']
#size = (5,2)
for a,b in zip(range(3),['gender','selfMade']): #zip() method takes iterable or containers and returns a single iterator object
    #lct = (0,a)
    #ax = plt.subplot2grid(size,lct)
    ax = plt.subplots(figsize=(5,3))
    data = df_bil[b].value_counts()
    labels = data.index
    x = data.values
    explode = (0, 0.2) #converts each element of the specified column(s) into a row
    angle = [45,80]
    fig = plt.pie(x=data, labels=labels, autopct='%1.1f%%',explode=explode,shadow=True,startangle=angle[a],colors=color)
    plt.title(b.capitalize(), size=10, color='black')

plt.show()
```

#### 2. Determine the distribution by age

```python
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={'height_ratios':(.2,.9)})
plt.subplots_adjust(hspace=0.5)

#Assigning a graph to each ax
#Box plot
sns.boxplot(df_bil['age'], ax=ax_box, color='darksalmon')
#Histogram
sns.histplot(df_bil['age'], ax=ax_hist, color='purple')
plt.axvline(df_bil['age'].mean(), color='lime', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(df_bil['age'].mean()*1.02, max_ylim*0.9, 'Mean: {:.0f}'.format(df_bil['age'].mean()))

plt.title(f'Age Distribution', horizontalalignment='center', color='black', fontsize=15)
plt.show()
```

#### 3. Determine the distribution of the top 25 countries, states, and cities

```python
plt.figure(figsize=(15,10))
data = df_bil['country'].value_counts()[:25]
x = data.index.tolist() #convert a series to list
y = data.values
bar = sns.barplot(y=x, x=y, palette='BuPu_r')
for n,i in enumerate(y):
    porc =f'{round((i/sum(df_bil["country"].value_counts().values))*100, 2)}%'
    bar.annotate(f'  {i} ({porc})', xy=(i,n), verticalalignment='center', horizontalalignment='left', fontsize=10)

bar.set_xlim([-10, 800]) #used to set the x-axis view limits
#Spines are the lines connecting the axis tick marks and noting the boundaries of the data area
bar.spines['top'].set_visible(False)
bar.spines['right'].set_visible(False)
#Set color of axis bars
bar.spines['left'].set_color('darkgrey')
bar.spines['bottom'].set_color('darkgrey')
plt.title('Top 25 Countries', fontsize=20, horizontalalignment='center', color='black', verticalalignment='bottom')
plt.yticks(fontsize=13)

plt.show()
```
